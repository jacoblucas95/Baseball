```python
import numpy as np
import pandas as pd
import datetime
```

```python
data = pd.read_csv('baseball.csv', low_memory=False)
```

```python
data.info()
```

```python
data.replace('NL','0', inplace=True)
```

```python
over_count = 0
under_count = 0
push_count = 0
for i,g in data.groupby(['gameno']):
    o_u = g[::2]
    if pd.to_numeric(o_u.totalruns.all()) > o_u.close_ou:
        over_count += 1
    elif pd.to_numeric(o_u.totalruns.all()) < o_u.close_ou:
        under_count += 1
    else:
        push_count += 1

        
print(over_count)
print(under_count)
print(push_count)
```
